```python
query_x = pd.read_sql_query(
                            """
                            SELECT Album.AlbumId As _ID_ALBUM , Album.title AS _TITLE_ALBUM , Artist.name AS _ARTIST_NAME
                            from Album , Artist
                            WHERE Album.artistid = Artist.artistid
                            LIMIT 10
                            """ , engine)
query_x
```

```python
query_y = pd.read_sql_query(
    """SELECT Album.albumid As _ALBUM_ID , Album.title as _ALBUM_Title , Artist.artistid As _ARTIST_ID , Artist.name As _ARTIST_NAME 
       from Album 
       INNER JOIN Artist
       on Artist.artistid = Album.artistid
       ORDER by Album.title desc
       LIMIT 10
    """ , engine)
query_y.sample()
```

```python
query_y.to_csv("./database/first_output.csv")
```

```python
query_z = pd.read_sql_query(
    """
    SELECT Album.AlbumId AS _ALBUM_ID , Album.title AS _ALBUM_Title , 
    Genre.genreid As _GENRE_ID , Genre.name As _GENRE_NAME , 
    MediaType.mediatypeid As _MEDIA_TYPE_ID , MediaType.name AS _MEDIA_TYPE_NAME , 
    Track.trackid As _TRACK_ID , Track.name AS _TRACK_NAME , Track.composer As _COMPOSER , 
    Track.milliseconds , Track.bytes , Track.unitprice , 
    PlaylistTrack.playlistid As _PLAY_LIST_ID ,
    InvoiceLine.invoicelineid , InvoiceLine.InvoiceId , InvoiceLine.unitprice , InvoiceLine.quantity
    from Track
    INNER JOIN Album
    on Album.albumid = Track.albumid
    INNER JOIN Genre
    on Genre.genreid = Track.genreid
    INNER JOIN MediaType
    on MediaType.mediatypeid = Track.mediatypeid
    INNER JOIN PlaylistTrack
    on Track.trackid = PlaylistTrack.TrackId
    INNER join InvoiceLine
    on Track.trackid = InvoiceLine.trackid
    ORDER by Album.Title desc
    """ , engine)
query_z.sample()
```

```python
query_z.to_csv("./database/second_output.csv")
```

```python
query_x1 = pd.read_sql_query(
       """
       SELECT Playlist.playlistid AS _PLAYLIST_ID , Playlist.name as _PLAYLIST_NAME , 
       PlaylistTrack.TrackId aS _TRACK_ID ,
       Track.name , Track.composer , Track.milliseconds , Track.bytes , Track.unitprice
       from PlaylistTrack
       INNER JOIN Playlist
       on PlaylistTrack.PlaylistId = Playlist.PlaylistId
       INNER join Track
       on PlaylistTrack.TrackId = Track.TrackId
       """ , engine)
query_x1.sample()
```

```python
query_x1.to_csv("./database/third_output.csv")
```

```python
f1 = pd.read_sql_query("""
SELECT  COUNT(Customer.customerid ) as _Number_OF_CUSTOMERS, Invoice.total 
from Customer
INNER join Invoice
on Customer.customerid = Invoice.customerid
GROUP by Invoice.total
HAVING  COUNT(Customer.customerid ) > 10
ORDER by COUNT(Customer.customerid ) desc
""" , engine)
f1.sample()
```

```python
f2 = pd.read_sql_query("""
select Customer.firstname , COUNT(Invoice.invoiceid) as _NUMBER_OF_INVOICES
from Customer
inner join Invoice
on Customer.customerid = Invoice.customerid
GROUP by Customer.firstname
HAVING _NUMBER_OF_INVOICES >= 7
ORDER by _NUMBER_OF_INVOICES desc
""" , engine)
f2.sample()
```

```python
f3 = pd.read_sql_query("""
SELECT total ,
CASE
    WHEN total > 5 THEN 'Good Invoice'
    WHEN total < 5 THEN 'bad Invoice'
END as _Comment_Text
from Invoice
""" , engine)
f3.sample()
```
